```python
# Importando bibliotecas
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

from IPython.display import display


from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression


#Configuracoes
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
# Funcoes utilizadas 
def missing_data(df):    
    missing_data = pd.DataFrame({'Tipo': df.dtypes,'Dados faltantes (%)': df.isna().sum() / df.shape[0]})
    missing_data.sort_values(by='Dados faltantes (%)', inplace=True, ascending=False)
    return missing_data


def sumary_df(df):
    print("Informações básicas do dataset")
    print("\nFormato:", df.shape)
    display(df.head(5))
    
    print("\nTipos de dados:")
    display(df.info())

    print("\nPercentual de dados faltantes:")
    display(missing_data(df).T)

    print("\nEstatísticas das features:")
    display(df.describe())
    
    
def calcula_erro(y_real, y_estimado):
    erro = abs(y_real.T - y_estimado)
    erro = erro.T

    # Substitui erros numericos por 100% de erro
    erro[erro == np.inf] = 100

    # Erro percentual médio
    erro_percent = (erro/len(y_real))*100
    erro_percent[erro_percent == np.inf] = 0
    
    return erro, erro_percent




def plot_resultados_validacao(y_test, y_modelo, metodo, nome_var):
    sns.set_style("white")
    fig, axes = plt.subplots(2, 2, figsize=(20, 12), dpi= 200, facecolor='w', edgecolor='k')
    fig.suptitle("Resultados validação: "+metodo)
    
    x_plot = np.arange(0, len(y_test))

    
    # Validacao
    plt.subplot(221)
    plt.title(nome_var+': Validação')
    
    plt.plot(x_plot, y_test, 'b', label='Real')
    plt.plot(x_plot, y_modelo ,'r', label=metodo)

    plt.ylabel(nome_var)
    plt.legend(loc='best')


    # Correlacao
    r2 = r2_score(y_test, y_modelo)
    
    min_grafico = min(min(y_test.T), min(y_pred_linear))
    # max_grafico = max(max(y_test.T), max(y_pred_linear))
    aux = pd.DataFrame(y_pred_linear)
    max_grafico = float(aux.quantile(0.75) + 3*aux.std())
    
    plt.subplot(222)
    plt.title(nome_var+': R² = {0:.2f}'.format(r2) )


    plt.plot(y_test, y_modelo, 'ko', alpha=0.2)
    plt.plot(x_plot, x_plot, 'b', label='Coef. R² = 1')
    plt.plot(x_plot, x_plot*r2 ,'r--', label='Coef. R² '+metodo)
    
    plt.xlim(0, max_grafico)
    plt.ylim(0, max_grafico)
    plt.ylabel(metodo)
    plt.xlabel('Medido')
    plt.legend(loc='best')

    # Calculando erro e erro percentual
    erro_abs, erro_perct = calcula_erro(y_test, y_modelo)
    
    # Erro absoluto  histograma
    plt.subplot(223)
    plt.title('Erro(%) '+ nome_var)
    sns.distplot( erro_perct, kde=False, color="b")
    plt.xlim(0,100)
    plt.ylabel('No. Amostras')
    plt.xlabel('Erro (%)')


    # RMSE
    rmse_modelo = np.sqrt(metrics.mean_squared_error(y_test, y_modelo))
    plt.subplot(224)

    plt.title('RMSE val. = {0:.2f}'.format(rmse_modelo) )
    
    plt.plot(x_plot, erro_abs, color="b", label='Erro absoluto')
    plt.plot(x_plot, np.ones(y_test.size)*rmse_modelo, color="r", label='+/- RMSE')
    plt.plot(x_plot, np.ones(y_test.size)*rmse_modelo*(-1), color="r")
    
    plt.ylabel('Erro absoluto')
    plt.legend(loc='best')


    plt.show()

# Quadro de estatisticas de modelo
def quadro_modelos(num_folds, y, idx_teste, mtx_pred):
    quadro = pd.DataFrame(0, index=["Média Erro (%)", "Média Erro abs", "Desv. Padrão Erro", "Coef. Determinação R²", "RMSE"],
                          columns=["Modelo 1", "Modelo 2", "Modelo 3", "Modelo 4", "Modelo 5"])

    for j in range(num_folds):
        # Definindo saida e previsto para cada modelo
        y_real = pd.DataFrame(y.iloc[idx_teste.iloc[:, j].dropna()])
        y_real.reset_index(inplace=True, drop=True)
        y_estimado = mtx_pred.iloc[:, j].dropna()

        # Calculando vetor de erro
        erro, erro_percent = calcula_erro(y_real, y_estimado)

        # Preenchendo o quadro
        quadro.iloc[0, j] = float(erro_percent.mean())
        quadro.iloc[1, j] = float(erro.mean())
        quadro.iloc[2, j] = float(erro.std())
        
        r2 = r2_score(y_real, y_estimado)
        quadro.iloc[3, j] = r2

        rmse = np.sqrt(metrics.mean_squared_error(y_real, y_estimado))
        quadro.iloc[4, j] = rmse

    # Preenchendo a média dos folds
    quadro["Média Modelos"] = quadro.mean(axis=1)
    return quadro



# Quadro Validação
def quadro_validacao(quadro, y_val, y_modelo):
    erro_val, erro_perc_val = calcula_erro(y_val, y_modelo)

    r2_val = r2_score(y_val, y_modelo)
    rmse_val = np.sqrt(metrics.mean_squared_error(y_val, y_modelo))

    quadro['Validação'] = [float(erro_perc_val.mean()), float(erro_val.mean()), float(erro_val.std()), r2_val, rmse_val]
    return quadro
```

## Separação dos dados de validação
Serão separados 10% dos dados, de forma aleatória, para que sejam usados como validação das métricas do modelo.

```python
# Reajustando formato do dataset para variavel de interesse ser a ultima coluna
aux = df['NU_NOTA_MT']
df.drop(columns='NU_NOTA_MT', inplace=True)
df['NU_NOTA_MT'] = aux
```

```python
df_val = df.sample(frac=0.1, replace=False, random_state=1)
df = df.drop(df_val.index)
print("Formato dados de validação:", df_val.shape)
print("Formato dados de treino/teste:", df.shape)
```

## Seperação de folds
Dados serão divididos em 5 folds, de forma aleatória, para que seja aplicada a técnica k-folds.

```python
X, y = df.iloc[:,0:-1], df.iloc[:,-1]
X_val, y_val = df_val.iloc[:,0:-1], df_val.iloc[:,-1]

num_folds = 5
kf = KFold(n_splits=num_folds, shuffle=True, random_state=123)

# dataframes que salvarão os índices das amostras de cada fold
idx_treino = pd.DataFrame(columns=['Treino_1', 'Treino_2', 'Treino_3', 'Treino_4', 'Treino_5'])
idx_teste = pd.DataFrame(columns=['Teste_1', 'Teste_2', 'Teste_3', 'Teste_4', 'Teste_5'])

i = 0

# Estabelecendo os folds
for train_index, test_index in kf.split(X):
    if(i == 0):
        idx_treino.iloc[:,i] = train_index
        idx_teste.iloc[:,i] = test_index
        
    # Ajustando formato do dataset caso a divisao de folds seja desigual
    if(len(train_index) > len(idx_treino) and i > 0):
        new_row = pd.DataFrame(np.nan, index=[len(train_index)], 
                                      columns=['Treino_1', 'Treino_2', 'Treino_3', 'Treino_4', 'Treino_5'])
        idx_treino = pd.concat([idx_treino, new_row])

    # Ajustando formato do novo fold caso a divisao de folds seja desigual    
    if(len(test_index) < len(idx_teste) and i > 0):
        test_index = np.append(test_index, np.nan)
        
    idx_teste.iloc[:,i] = test_index
    idx_treino.iloc[:,i] = train_index
    i = i+1
```

## Aplicação do modelo de regressão linear

```python
# Treinando modelos de regressao linear

# Criando matriz de resultados
mtx_pred_linear = pd.DataFrame(0, index=np.arange(len(idx_teste)), 
                               columns=["Modelo_1", "Modelo_2", "Modelo_3", "Modelo_4", "Modelo_5"])

# Treianando modelos
for i in range(num_folds):
    X_train = X.iloc[idx_treino.iloc[:, i].dropna(), :]
    y_train = y.iloc[idx_treino.iloc[:, i].dropna()]
    
    regressor = LinearRegression()  
    regressor.fit(X_train, y_train)

    X_test = X.iloc[idx_teste.iloc[:, i].dropna(), :]
    y_pred_linear = regressor.predict(X_test)
    
    if(len(y_pred_linear) < len(mtx_pred_linear)):
        y_pred_linear = np.append(y_pred_linear, np.nan)
    
    # Adicionando y_previsto na matriz de resultados
    mtx_pred_linear.iloc[:, i] = y_pred_linear
```

### Dados de treinamento / teste

```python
# Estatisticas
qd_reg_linear = quadro_modelos(num_folds, y, idx_teste, mtx_pred_linear)
display(qd_reg_linear)
```
