# Descubra quem fez o ENEM 2016 apenas para treino

- Escolha de variáveis para análise
- Analisando os candidatos treineiros
- Modelo de classificação



```python
# Importando bibliotecas
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_confusion_matrix

from sklearn.metrics import classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
#Configuracoes
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)


import warnings
warnings.filterwarnings("ignore")
```

```python
# Funcoes utilizadas 
def missing_data(df):    
    missing_data = pd.DataFrame({'Tipo': df.dtypes,'Dados faltantes (%)': df.isna().sum() / df.shape[0]})
    missing_data.sort_values(by='Dados faltantes (%)', inplace=True, ascending=False)
    return missing_data

def sumary_df(df):
    print("Informações básicas do dataset")
    print("\nFormato:", df.shape)
    display(df.head(5))
    
    print("\nTipos de dados:")
    display(df.info())

    print("\nPercentual de dados faltantes:")
    display(missing_data(df).T)

    print("\nEstatísticas das features:")
    display(df.describe())
```

```python
df_treino = pd.read_csv('train.csv')
df_teste = pd.read_csv('test.csv')
```

```python
print("Formato dados de treino: ", df_treino.shape)
df_treino.head()
```

# Escolha de variáveis para análise

A escolha inicial de variáveis será baseada em quais features estão disponíveis no dataset de teste.

```python
print("Formato dados de teste: ", df_teste.shape)
```

```python
features = df_teste.columns
df = df_treino.loc[:, features]
df['IN_TREINEIRO'] = df_treino['IN_TREINEIRO']
print("Formato dados treino: ", df.shape)
```

# Analisando os candidatos treineiros
## Dataset

```python
sumary_df(df)
```

## Comparação:  Base total vs Treineiros

Serão comparadas as variáveis, onde há a hipótese de serem as mais importanets para o problema: `NU_IDADE`, `TP_ST_CONCLUSAO`, `TP_ANO_CONCLUIU`. 

Significado da variável `TP_ST_CONCLUSAO`:
1. Já concluí o Ensino Médio
-	Estou cursando e concluirei o Ensino Médio em 2016
-	Estou cursando e concluirei o Ensino Médio após 2016
-	Não concluí e não estou cursando o Ensino Médio

Significado da variável `TP_ANO_CONCLUIU`:
0.	Não informado
1.	2015
2.	2014
3.	2013
4.	2012
5.	2011
6.	2010
7.	2009
8.	2008
9.	2007
10.	Anterior a 2007

```python
features_compare = ['NU_IDADE', 'TP_ST_CONCLUSAO', 'TP_ANO_CONCLUIU']

df_treineiro = df.copy()
df_treineiro = df_treineiro[df_treineiro['IN_TREINEIRO'] == 1]
```

```python
fig, ax = plt.subplots(3, 2, figsize=(15, 5), facecolor='w', edgecolor='k')
plt.suptitle("Comparação distribuições das variáveis")

ordem = 321
for i in features_compare:
    plt.subplot(ordem)
    if (ordem == 321):
        plt.title("Base total")
    sns.distplot(df[i], color='blue', label=i)
    plt.legend()
    plt.xlabel("")
    ordem = ordem + 1
    
    plt.subplot(ordem)
    if (ordem == 322):
        plt.title("Treineiros")
    sns.distplot(df_treineiro[i], color='gray', label=i)
    plt.legend()
    plt.xlabel("")
    ordem = ordem + 1
    
    
plt.show()
```

Nota-se que estas três variáveis possuem uma alta relação com o problema abordado.

# Modelo de classificação: Regressão logística

```python
features_modelo = ['NU_IDADE', 'TP_ST_CONCLUSAO', 'TP_ANO_CONCLUIU']
X = df.loc[:, features_modelo]
y = df['IN_TREINEIRO']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

logistic = LogisticRegression()

logistic.fit(X_train, y_train)
```

```python
# Contrução da matriz de confusão
y_previsto = logistic.predict(X_test)
matriz_confusao = confusion_matrix(y_test, y_previsto)

# Realiza o plot da matriz de confusão
fig, ax = plot_confusion_matrix(conf_mat=matriz_confusao)
plt.show()
```

# Resposta final: score de 99.03%

```python
X = df_teste.loc[:, features_modelo]
predicao = logistic.predict(X)
predicao = pd.Series(predicao)

answer = pd.DataFrame(df_teste['NU_INSCRICAO'])
answer['IN_TREINEIRO'] = predicao

answer.to_csv("answer.csv", index=False)
```
